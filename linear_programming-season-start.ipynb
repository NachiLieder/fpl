```python
df_sorted['PPM'] = df_sorted['total_points']/df_sorted['minutes']

df_sorted['PPM']=(df_sorted['PPM']-df_sorted['PPM'].mean())/df_sorted['PPM'].std()
df_sorted['points_per_game']=(df_sorted['points_per_game']-df_sorted['points_per_game'].mean())/df_sorted['points_per_game'].std()
df_sorted['total_points']=(df_sorted['total_points']-df_sorted['total_points'].mean())/df_sorted['total_points'].std()


df_sorted['aggregated']= df_sorted['PPM'] + df_sorted['points_per_game'] + df_sorted['total_points']
df_sorted['aggregated_std']= df_sorted[['PPM','points_per_game','total_points']].std(axis=1)

# df_sorted=df_sorted[df_sorted['aggregated']>1].reset_index(drop=True)
```

As a first example, consider the solution of the 0/1 knapsack problem: given a set 𝐼 of items, each one
with a weight 𝑤𝑖 and estimated profit 𝑝𝑖
, one wants to select a subset with maximum profit such that the
summation of the weights of the selected items is less or equal to the knapsack capacity 𝑐. Considering a
set of decision binary variables 𝑥𝑖 that receive value 1 if the 𝑖-th item is selected, or 0 if not, the resulting
mathematical programming formulation is:

```python
from mip import Model, xsum, maximize, BINARY

# wi = weights
# pi = profit
# c = capacity : total cost (100)


p = list(val.values)
w = list(costs.values)
std_penalty = list(std.values)
c= 
I = range(len(w))
I_g = range(len(gk))
I_d= range(len(defense))
I_m= range(len(mid))
I_f= range(len(fwd))

m = Model("knapsack")
# define vector of chosen players
x = [m.add_var(var_type=BINARY) for i in I]

# objective function
m.objective = maximize(xsum((p[i] * x[i])-np.log(std_penalty[i]) for i in I) )

# cost constraint
m += xsum(w[i] * x[i] for i in I) <= c

#constraint for number of GK
m += xsum(x[i] for i in I_g )==1
#constraint for number of DEF
m += xsum(x[i + len(gk)] for i in I_d )==5
#constraint for number of MID
m += xsum(x[i+ len(gk)+ len(defense)] for i in I_m )==5
#constraint for number of FWD
m += xsum(x[i+ len(gk)+ len(defense)+ len(mid)] for i in I_f )==2

# exclude players by id
# example lloris = 1
ids_to_exclude = [173,123,9,103,61,177,176]

m += xsum(x[i] for i in ids_to_exclude) == 0


m.optimize()

selected = [i for i in I if x[i].x >= 0.99]
print("selected items: {} ".format(selected))


# for i in selected:
#     print
df_sorted.iloc[selected].sort_values('singular_name_short')
```
